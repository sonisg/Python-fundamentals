```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#to create a array 
a=np.arange(1,10)
a
```

```python
b=np.arange(11,20)
b
```

```python
#plot the data a and b
plt.scatter(a,b,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph')
plt.show()
```

```python
#for an exponential plot
y=a*a
y
```

```python
plt.plot(a,y,'r*',linestyle='dashed',linewidth=2, markersize=10)
plt.show()
```

```python
#creating subplots
plt.subplot(2,2,1)
plt.plot(a,y,'r--')
plt.subplot(2,2,2)
plt.plot(a,y,'g*--')
plt.subplot(2,2,3)
plt.plot(a,y,'bo')
plt.subplot(2,2,4)
plt.plot(a,y,'go')
plt.show()
```

```python
#To create a line using equation
x=np.arange(1,11)
y=3*x+6
x
```

```python
y
```

```python
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
#SINE GRAPH
```

```python
x=np.arange(0,4*np.pi,0.1)
x
```

```python
y=np.sin(x)
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show()
```

```python
#to plot a sine and cosine graph
x=np.arange(0,5*np.pi,0.1)
s=np.sin(x)
c=np.cos(x)
plt.subplot(2,2,1)
plt.plot(x, s,'r--') 

plt.subplot(2,2,2)
plt.plot(x, c,'g--') 

plt.show()
```

```python
#BAR PLOT
x=[4,6,3,5]
y=[8,6,1,9]
plt.bar(x, y) 
x1=[2,5,6,1,3]
y1=[1,8,5,1,2]
plt.bar(x1,y1,color='g')
plt.show()
```

```python
#HISTOGRAMS
a=np.array([2,3,6,8,11,13,14,15])
plt.hist(a)
plt.show()
```

```python
# Data to plot -Piechart PLot
labels = 'Python', 'C++', 'R', 'Java'
sizes = [155, 230, 145, 210]
colors = ['gold', 'green', 'coral', 'blue']
explode = (0.1, 0.3, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
