```python
import seaborn as sns
```

```python
#IRIS DATASET
df=sns.load_dataset("tips")
df.head()
```

```python
#Correlation with Heatmap
```

```python
df.corr()
```

```python
#A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. 
#The values of the first dimensions appear as rows of the table, while the values of the second dimension are represented by the columns of the table.
```

```python
sns.heatmap(df.corr())
```

```python
#JoinPlot
#It is used to draw a plot of two variables with bivariate and univariate graphs. It basically combines two different plots.
#Syntax ---- jointplot(x, y[, data, kind, stat_func, ...]) 
```

```python
#UNIVARIATE ANALYSIS
#“Uni” means “one”, so in other words your data has only one variable. 
#It doesn't deal with causes or relationships (unlike regression ) and it's major purpose is to describe; 
#It takes data, summarizes that data and finds patterns in the data
```

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = df, kind ='kde') 
# KDE shows the density where the points match up the most 
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
#Pair plot
#It represents pairwise relation across the entire dataframe and supports an additional argument called hue for categorical separation. 
#syntax ---- pairplot(data[, hue, hue_order, palette, …])  
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
#DIST PLOT
#It is used basically for univariant set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset.
#syntax ---- distplot(a[, bins, hist, kde, rug, fit, ...])
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#RUG PLOT
#It plots datapoints in an array as sticks on an axis.Just like a distplot it takes a single column. Instead of drawing a histogram it creates dashes all across the plot.
#syntax ---- rugplot(a[, height, axis, ax])  
```

```python
sns.rugplot(df['total_bill']) 
```

```python
#CATEGORIAL PLOTS
```

```python
### Count plot
#A countplot basically counts the categories and returns a count of their occurrences. 
#Syntax ---- countplot([x, y, hue, data, order, …])
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#BAR PLOT
#A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean
#Syntax --- barplot([x, y, hue, data, order, hue_order, …])
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
#BOX PLOT
#A boxplot is sometimes known as the box and whisker plot.It shows the distribution of the quantitative data that represents the comparisons between variables
#Syntax ---- boxplot([x, y, hue, data, order, hue_order, …])
```

```python
sns.boxplot(x ='day', y ='total_bill', data = df, hue ='smoker') 
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#VIOLIN PLOT
#It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution.
#Syntax --- violinplot([x, y, hue, data, order, …])
```

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```

```python
sns.violinplot(x ='day', y ='total_bill', data = df, hue ='sex', split = True) 
```

```python
#Strip plot
#It basically creates a scatter plot based on the category.
#syntax ---- stripplot([x, y, hue, data, order, …])
```

```python
sns.stripplot(x ='day', y ='total_bill', data = df,  
              jitter = True, hue ='smoker', dodge = True) 
```

```python
#SWARM PLOT
#It is very similar to the stripplot except the fact that the points are adjusted so that they do not overlap.
```

```python
sns.swarmplot(x ='day', y ='total_bill', data = df)
```

```python
sns.violinplot(x ='day', y ='total_bill', data = df) 
sns.swarmplot(x ='day', y ='total_bill', data = df, color ='black') 
```

```python
#FACTOR PLOT
#It is the most general of all these plots and provides a parameter called kind to choose the kind of plot we want thus saving us from the trouble of writing these plots separately. 
#The kind parameter can be bar, violin, swarm etc.
```

```python
sns.factorplot(x ='day', y ='total_bill', data = df, kind ='bar') 
```
